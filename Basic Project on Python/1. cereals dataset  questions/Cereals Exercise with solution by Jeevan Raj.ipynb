#### 10. Add new column as HL_Calories where more than average calories should be categorized as  HIGH and less than average calories should be categorized as LOW 

```python
df.head()
```

```python
import numpy as np
```

```python
df['HL_Calories'] = np.where(df.calories>df.calories.mean(),'HIGH','LOW')
```

```python
df.head()
```

```python
df.drop('HL_Calories',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['HL_Calories'] = df['calories'].apply(lambda x : 'HIGH' if x > df.calories.mean()  else 'LOW')
df.head()
```
